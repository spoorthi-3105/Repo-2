```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv("/Sample - Superstore.csv", encoding='latin-1')
df.head(5)
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
df_clean = df.dropna()
```

```python
plt.figure(figsize=(10,6))
numeric_df = df.select_dtypes(include=['float64', 'int64'])
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Category', data=df)
plt.title('Category Distribution')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Sub-Category', data=df)
plt.title('Sub-Category Distribution')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Region', y='Sales', data=df)
plt.title('Sales by Region')
plt.show()
```

```python
print("Top 10 Products by Sales:")
print(df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10))
```

```python
print("Top 10 Products by Profit:")
print(df.groupby('Product Name')['Profit'].sum().sort_values(ascending=False).head(10))
```

```python
fig = px.bar(df.groupby('Category')['Sales'].sum().reset_index(), x='Category', y='Sales', title='Sales by Category')
fig.show()

fig = px.bar(df.groupby('Region')['Sales'].sum().reset_index(), x='Region', y='Sales', title='Sales by Region')
fig.show()
```

```python
corr_matrix = df[['Sales', 'Quantity', 'Discount', 'Profit']].corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
customer_segments = df.groupby('Region')['Sales'].sum().reset_index()
fig = px.pie(customer_segments, values='Sales', names='Region', title='Customer Segments by Region')
fig.show()
```

```python
top_products_sales = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
top_products_profit = df.groupby('Product Name')['Profit'].sum().sort_values(ascending=False).head(10)

# Calculate profit by region
region_profit = df.groupby('Region')['Profit'].sum().sort_values(ascending=False)

print("Key Takeaways:")
print("1. Top-performing products (by sales): ", top_products_sales.index[:3].tolist())
print("2. Most profitable region: ", region_profit.index[0])
print("3. Customer segments by region (based on sales): ", customer_segments['Region'].unique().tolist())
```
